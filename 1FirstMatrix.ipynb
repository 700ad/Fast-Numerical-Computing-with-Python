```python
def checkSal(players):
    Color = {"DeMarDeRozan":"Green","StephenCurry":"Red","LeBronJames":"Blue","NikolaJokic":"Orange","BogdanBogdanovic":"Black"}
    Marker = {"DeMarDeRozan":"o","StephenCurry":"s","LeBronJames":"*","NikolaJokic":"d","BogdanBogdanovic":"o"}
    for name in players:
        plt.plot(Salary[Pdict[name]], c=Color[name], ls='--', marker=Marker[name],ms=5,label=name)
        plt.legend(loc='upper left',bbox_to_anchor=(1,1))
        plt.xticks(list(range(0,5)), Seasons)
    plt.show()
```
